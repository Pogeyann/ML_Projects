```python
#Basic data exploration

import pandas as pd
melb = pd.read_csv('D:/TUTORIALS/Datas/data_zip/archive/melb_data.csv',index_col=0)
melb

```

```python
melb.describe()
#mean is the average
# standard deviation measures how numerically spread out values are
```

```python

melb.columns
```

```python

melb.isna().sum()

melb = melb.dropna(axis=0)

melb.isna().sum()
```

```python


y = melb['Price'] #melb.Price
y
```

```python

# Features:columns that are inputted into ur model 
# (and later used to make predictions) are called features '''
x = melb[['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']]
x
```

```python

x.describe()
```

```python

x.head()
```

```python

##Building your model

from sklearn.tree import DecisionTreeRegressor
#Defining model
melb_model = DecisionTreeRegressor(random_state= 1)
#Fit model
melb_model.fit(x,y)
```

```python


print('making predictions for the following 5 houses:')
print(x.head())
print('the predictions are')
print(melb_model.predict(x.head()))
```
