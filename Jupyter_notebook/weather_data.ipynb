## Model Classifications

```python
wee['Date/Time']=pd.to_datetime(wee['Date/Time'])
#extracting the year and convert it to a numeric representation
wee['year'] = wee['Date/Time'].dt.year.astype(int)
#extracting the onth and convert it ot a numeric representation
wee['month'] = wee['Date/Time'].dt.month.astype(int)
#extracting the day and convert it to a numeric representation
wee['day'] = wee['Date/Time'].dt.day.astype(int)
wee['time'] = wee['Date/Time'].dt.hour.astype(int)
#re-indexing
wee = wee.reindex(columns=['Temp_C', 'Dew Point Temp_C', 'Rel Hum_%', 'Wind Speed_km/h',
       'Visibility_km', 'Press_kPa', 'year', 'month', 'day',
       'time','Weather']
```

```python
wee.drop('Date/Time', axis=1)
```
