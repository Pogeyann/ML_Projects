```python
img_grey=cv2.imread('/home/mush/Computer_vision/opencv/istockphoto-1211345565-612x612.jpg',cv2.IMREAD_GRAYSCALE)

```

```python
cv2.imshow('grayscale image',img_grey)
cv2.waitKey() #set time 
cv2.destroyAllWindows() #Closes windows
```

```python
#Binary image- THreshold image

(thresh,binary_img)=cv2.threshold(img_grey,0,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)
```

```python
cv2.imshow('binary image',binary_img)
#cv2.imshow('original image',img)
cv2.imshow('grayscale image',img_grey)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
edge_img = cv2.Canny(img,255,0)
```

```python
cv2.imshow('original image',img)
cv2.imshow('edge image',edge_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
```

```python
converted_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#converted to bgr to rgb .otherwise bgr colors only ,converts bluish tone image
```

```python
plt.imshow(converted_img)
```

```python
#for creating a rectangle
converted_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
rect = cv2.rectangle(converted_img,pt1=(400,50),pt2=(150,300),color=(204,92,90),thickness=50)#negative value give s inside color
plt.imshow(rect)
```

```python
converted_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
circl = cv2.circle(converted_img,center=(300,150),radius=100,color=(204,92,90),thickness=-1)
plt.imshow(circl)
```

```python
converted_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
puttext = cv2.putText(converted_img,text='luminar',org=(100,300),fontFace=cv2.FONT_HERSHEY_SCRIPT_COMPLEX,fontScale=5,color=(0,255,2))
plt.imshow(puttext)
```
