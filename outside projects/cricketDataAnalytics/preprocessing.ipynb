```python
with open('D:/vscode/outside projects/cricketDataAnalytics/codebasics/t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)  

df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.rename(columns={'scorecard':'match_id'},inplace=True)
```

```python
with open('D:/vscode/outside projects/cricketDataAnalytics/codebasics/t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)

    all_records = []

    for rec in data:
        all_records.extend(rec['battingSummary'])

all_records

df_batting = pd.DataFrame(all_records)
df_batting.head()

    
```

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: 'out' if len(x) > 0 else 'notout' )
```

```python
df_batting.drop('dismissal',axis=1,inplace=True)
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€',''))
```

```python
match_ids_dict = {}
for index,row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']

    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
match_ids_dict
```

```python
df_batting['mach_id'] = df_batting['match'].map(match_ids_dict)
```

```python
df_batting.rename({'mach_id':'match_id'},axis=1,inplace=True)
```
