## Eye Detection

```python
eye_cascade = cv2.CascadeClassifier('/home/mush/Computer_vision/face_detection/haarcascade_eye.xml')
```

```python
img1 = cv2.imread('/home/mush/Computer_vision/face_detection/mohanlal.jpeg')
```

```python
converted_img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
```

```python
gray1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
```

```python
eyes = eye_cascade.detectMultiScale(gray1,scaleFactor=1.1,minNeighbors=4)
```

```python
len(eyes)
```

```python
for(x,y,w,h) in eyes:
    cv2.rectangle(converted_img1,(x,y),(x+w,y+h),(255,0,0),3)
```

```python
plt.imshow(converted_img1)
```

## Eye video Detection

```python
video = cv2.VideoCapture('/home/mush/Computer_vision/videoplayback.mp4')
while True:
    success,frame = video.read()

    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    face = face_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=4)

    eyes= eye_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=4)
    #print(len(faces))
    #print(len(eyes))
    for x,y,w,h in face:
        cv2.rectangle(frame,(x,y),(x+w,y+h),color=(0,255,0),thickness=3)
#Video including
    for x,y,w,h in eyes:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,240),2)



    cv2.imshow('video',frame)
    if cv2.waitKey(1) & 0xFF==27:
        break

video.release()
cv2.destroyAllWindows()
```
