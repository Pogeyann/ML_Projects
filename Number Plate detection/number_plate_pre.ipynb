```python
plate_cascade=cv2.CascadeClassifier("/home/mush/Computer vision/Number Plate detection/haarcascade_russian_plate_number.xml")
img=cv2.imread("/home/mush/Computer vision/Number Plate detection/vehicle5.jpeg")
img.shape
# img=cv2.resize(img,(720,720))
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
no_plate=plate_cascade.detectMultiScale(gray_img,scaleFactor=1.3,minNeighbors=4)
print(len(no_plate))
```

```python

img=cv2.imread("/home/mush/Computer vision/Number Plate detection/vehicle5.jpeg")

for(x,y,w,h) in no_plate:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    # plt.imshow(img)
    crop=gray_img[y:y+h,x:x+w]
    # cv2.imshow('plate',crop)
    # cv2.waitKey()
    # cv2.destroyAllWindows()
    text = pytesseract.image_to_string(crop)
    new_text="".join(ch for ch in text if ch.isalnum())
    
    cv2.rectangle(img,(0,0),(800,200),color=(255,255,255),thickness=-1)
    cv2.putText(img,new_text, (50,50),cv2.FONT_HERSHEY_DUPLEX,2,(0,0,0),3)
    print(new_text)
    for i in dic:
        if new_text[0:4]==i:
            cv2.putText(img,dic[i], (50,150), cv2.FONT_HERSHEY_DUPLEX,1,(0,0,0),2)
            print(dic[i])
    
# plt.imshow(img)
cv2.namedWindow("resulted_img", cv2.WINDOW_NORMAL)  
cv2.imshow('resulted_img',img)
cv2.waitKey()
cv2.destroyAllWindows()
```
